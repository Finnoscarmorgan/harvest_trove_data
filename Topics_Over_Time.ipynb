```python
"""

This program loads in the file 'whole_data_with_topics.csv' and creates a new dataframe that counts the number of times each topic (from the 'topic' column) appears in each year (using the 'year' column).
This information is then visualise din a line chart for each topic for each year.

"""

#%%
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import re

#%%

# Load in the data
df = pd.read_csv('LOAD CSV FILE HERE')

```

```python
#transform the 'year' column into an integer
df['year'] = df['year'].astype(int)
```

```python
# Create a new dataframe that counts the number of times each topic appears in each year
df2 = df.groupby(['year','topic']).size().reset_index(name='counts')
```

```python
# visualise the data in a line chart for each topic for each year
sns.set_style("darkgrid")
sns.set_context("poster")
sns.set_palette("bright")
sns.set(font_scale=1.5)
sns.relplot(x="year", y="counts", hue="topic", kind="line", data=df2)
plt.xticks(rotation=45)
plt.show()

#%%
```

```python
#visualise the data in a line chart for each topic for each year with a different topic on each subplot (facet)
sns.set_style("darkgrid")
sns.set_context("poster")
sns.set_palette("bright")
sns.set(font_scale=1.5)
sns.relplot(x="year", y="counts", hue="topic", kind="line", col="topic", col_wrap=3, data=df2)
plt.xticks(rotation=45)
plt.show()
```
